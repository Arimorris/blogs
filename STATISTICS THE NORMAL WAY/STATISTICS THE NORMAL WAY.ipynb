```python
# cancel data from personal project being worked on with flight information
cancel = pd.read_csv(cwd + '/cancel.csv')
```

```python
dept = cancel['CRS_DEP_TIME'].apply(lambda x: round(x/100,2))
```

```python
pos = 0
for d in dept:
    dec = (math.modf(d))
    if round(dec[0], 2) >= .29:
        dept[pos] = int(dept[pos]) +1 
    else:
        dept[pos] = int(dept[pos])
    pos +=1
```

```python
hour = [x for x in range(25)]
time = dept.value_counts()
timefreq = [time[x] for x in range(25)]
```

```python
plt.figure(figsize=(20,10))
plt.bar(hour, timefreq)
plt.xticks(range(0,25))
plt.show()
```

## NORMAL DISTRIBUTION

```python
values= np.random.normal(70,5, 1000)
count, bins, ignored = plt.hist(values, normed =True, color ='red')

plt.plot(bins, 1/(5 * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - 70)**2 / (2 * 5**2) ),       linewidth=3, color='y')
plt.show()
```

```python
mn, std1, mn2, std2 = 40, 15, 15, 2

plt.figure(figsize=(10,5))
values= np.random.normal(mn,std1, 10000)
count, bins, ignored = plt.hist(values, normed =True, color ='orange')

plt.plot(bins, 1/(std1 * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mn)**2 / (2 * std1**2) ),       linewidth=3, color='black')

values= np.random.normal(mn2,std2, 10000)
plt.hist(values, normed =True, color ='purple')

plt.plot(bins, 1/(std2 * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mn2)**2 / (10 * std2**2) ),       linewidth=3, color='black')
plt.show()
```
